```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv(r"C:\Users\mohit\Downloads\Daily_data_of_reservoir_level_of_Central_Water_Commission_(CWC)_Agency_during_April_2023.csv")
df.head()
```

```python
df_new=df.dropna()
df_new
```

```python
plt.boxplot(df['Live_capacity_FRL'])
plt.title('plot of Live_capacity_FRL')

```

```python
plt.scatter(df_new.Level,df_new.Storage)
plt.xlabel('Level')
plt.ylabel('Storage')
plt.title('Storage vs Level')
```

```python
import numpy as np
bandwidth=0.25
fig, ax=plt.subplots(figsize=(30,200))
reservoir_names=df_new.Reservoir_name
level1=df_new.Level
level2=df_new.Full_reservoir_level
pos1=np.arange(len(reservoir_names))
pos2=[x + bandwidth for x in pos1]
ax.barh(pos1,level1,bandwidth,label='Current Level')
ax.barh(pos2,level2,bandwidth,label='Full or Maximum Resevior level')
```

```python
ax.invert_yaxis()
for i ,(val1,val2) in enumerate(zip(level1,level2)):
    plt.text(val1+0.2,pos1[i]+0.5,str(round(val1,2)),fontsize=20,fontweight=20)
    plt.text(val2+0.2,pos2[i]+0.5,str(round(val2,2)),fontsize=20,fontweight=20)
    ax.set_title('reservoir name and thier level',loc='left')
    ax.set_yticks((pos1+pos2)/2)
    ax.set_yticklabels(reservoir_names)
    ax.legend(loc='upper right',title='Legend')
    plt.show()
```
